```python
import numpy as np
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/dineshpiyasamara/LaptopPricePredictor/master/model%20building/laptop_price.csv"
data = pd.read_csv(url, encoding='latin-1')
data.head()
```

```python
data.info()
```

```python
data.head(2)
```

```python
data['Ram'] = data['Ram'].str.replace('GB','').astype('int32')
data['Weight'] = data['Weight'].str.replace('kg','').astype('float32')
```

```python
data.corr()['Price_euros']
```

```python
data['Company'].value_counts()
```

```python
len(data['Company'].value_counts())
```

```python
def add_company(inpt):
    if inpt == 'Samsung' or inpt == 'Razer' or inpt == 'Mediacom' or inpt == 'Microsoft' or inpt == 'Xiaomi' or inpt == 'Vero' or inpt == 'Chuwi' or inpt == 'Google' or inpt == 'Fujitsu' or inpt == 'LG' or inpt == 'Huawei':
        return 'Other' 
    else: 
        return inpt
data['Company'] = data['Company'].apply(add_company)
```

```python
data['Touchscreen'] = data['ScreenResolution'].apply(lambda x:1 if 'Touchscreen' in x else 0) 
data['Ips'] = data['ScreenResolution'].apply(lambda x:1 if 'IPS' in x else 0) 
```

```python
data.head(2)
```

```python
data['Cpu'].value_counts()
```

```python
data['cpu_name'] = data['Cpu'].apply(lambda x:" ".join(x.split()[0:3]))
```

```python
data['cpu_name'].value_counts()
```

```python
def set_processor(name):
    if name == 'Intel Core i7' or name == 'Intel Core i5' or name == 'Intel Core i3': 
        return name
    else:
        if name.split()[0] == 'AMD':
            return 'AMD'
        else:
            return 'Other'
data['cpu_name'] = data['cpu_name'].apply(set_processor)
```

```python
data['Gpu'].value_counts()
```

```python
len(data['Gpu'].value_counts())
```

```python
data['gpu_name'] = data['Gpu'].apply(lambda x:" ".join(x.split() [0:1]))
```

```python
data['gpu_name'].value_counts()
```

```python
data.shape
```

```python
data = data[data['gpu_name'] != 'ARM']
```

```python
data.shape
```

```python
data.head(2)
```

```python
data['OpSys'].value_counts()
```

```python
def set_os(inpt):
    if inpt == 'Windows 10' or inpt == 'Windows 7' or inpt == 'Windows 10 S': 
        return 'Windows'
    elif inpt == 'macOS' or inpt == 'Mac OS X':
        return 'Mac'
    elif inpt == 'Linux':
        return inpt
    else:
        return 'Other'
data['OpSys'] = data['OpSys'].apply(set_os)
```

```python
data['OpSys'].value_counts()
```

```python
data.head(2)
```

```python
data = data.drop(columns=['laptop_ID', 'Inches', 'Product', 'ScreenResolution', 'Cpu', 'Gpu'])
```

```python
data.head(2)
```

```python
data = pd.get_dummies(data)
```

```python
data.head(2)
```

```python
data.shape
```

```python
X = data.drop('Price_euros', axis=1)
y = data['Price_euros']
```

```python
#!pip install scikit-learn
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25) # Splitting dataset into training and testing sets.
```

```python
X_train.shape, X_test.shape
```

```python
def model_acc(model):
    model.fit(X_train, y_train)
    acc = model.score(X_test, y_test)
    print(str(model)+' --> '+str(acc)) # Model accuracy evaluation function
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression() 
model_acc(lr)  # Linear regression model accuracy calculation.

from sklearn.linear_model import Lasso 
lasso = Lasso() 
model_acc(lasso) # Lasso model accuracy calculation.

from sklearn.tree import DecisionTreeRegressor 
dt = DecisionTreeRegressor()
model_acc(dt) # DecisionTreeRegressor model accuracy calculation. 

from sklearn.ensemble import RandomForestRegressor 
rf = RandomForestRegressor() 
model_acc(rf) # Random ForestRegressor n model accuracy calculation.
```

```python
# Topic: Importing necessary modules
from sklearn.model_selection import GridSearchCV  # Import GridSearchCV for hyperparameter tuning

# Topic: Defining hyperparameter grid
parameters = {'n_estimators': [10, 50, 100],  # Define the number of trees in the forest
              'criterion': ['squared_error', 'absolute_error', 'poisson']}  # Define the splitting criterion

# Topic: Creating GridSearchCV object
grid_obj = GridSearchCV(estimator=rf, param_grid=parameters)  # Create a GridSearchCV object with random forest model and parameter grid

# Topic: Fitting GridSearchCV object to training data
grid_fit = grid_obj.fit(X_train, y_train)  # Fit the GridSearchCV object to the training data to find the best model

# Topic: Extracting best model from GridSearchCV results
best_model = grid_fit.best_estimator_  # Get the best model found by GridSearchCV
best_model  # Display the best model
```

```python
best_model.score(X_test, y_test) # Evaluating best model's accuracy on test data
```

```python
# Topic: Saving the Best Model
import pickle  # Import pickle module for object serialization

# Open a file to save the best model
with open('predictor.pickle', 'wb') as file:
    # Dump the best model into the file using pickle
    pickle.dump(best_model, file)
```
